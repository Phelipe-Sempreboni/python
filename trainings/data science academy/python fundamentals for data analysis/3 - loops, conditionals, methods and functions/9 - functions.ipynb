### Funções

```python
# Definindo uma função
# Essa função só irá imprimir na tela a frase Hello Word.

def primeiraFunc():
    print('Hello Word')
```

```python
# Executando a função primeiraFunc que irá imprimir na tela a frase Hello Word

primeiraFunc()
```

```python
# Definindo uma função com paramêtro, ou seja, necessita que seja informado um paramêtro para que a função o retorne
# Notar que é quase a mesma função acima, porém, possuí paramêtro com um placeholder

def primeiraFunc(nome):
    print('Hello %s ' %(nome))
```

```python
# Executando a função primeiraFunc que irá imprimir na tela a frase Hello + paramêtro, que neste caso é Aluno

primeiraFunc('Aluno')
```

```python
# Definindo uma função com um loop for e objeto range, de 0 a 5, que irá imprimir a frase Número + (i), ou seja, itens do loop

def funcLeitura():
    for i in range(0,5):
        print('Número ' + str(i))
```

```python
# Executando a função funcLeitura

funcLeitura()
```

```python
# Função para realizar a soma de números

def addNum(firstnum, secondnum):
    print('Primeiro número: ' + str(firstnum))
    print('Segundo número: ' + str(secondnum))
    print('Soma: ', firstnum + secondnum)
```

```python
# Executando a função addNum

addNum(25, 25)
```

### Variáveis locais e globais

```python
# Variável global
# Variável global é uma variável que pode ser utilizada em qualquer parte do código, ou seja, ela não é especifica de um bloco
# Neste caso o bloco referido é o da definição da função

var_global = 10 # Essa variável é global

def multiply(num1, num2):
    var_global = num1 * num2 # Essa variável é local
    print(var_global) # Será impresso na tela o valor da variável global
```

```python
# Executando a função multiply

multiply(3,3)
```

```python
# Notar que se executarmos a var_global, será executada que possuí o valor 10
# Não será executado a var_global da função, pois, essa variável é local, ou seja, é somente do bloco da função

print(var_global)
```

```python
# Variável global
# Variável global é uma variável que pode ser utilizada em qualquer parte do código, ou seja, ela não é especifica de um bloco
# Neste caso o bloco referido é o da definição da função
# Neste bloco diferenciamos a var_global da var_local, logo, fica fácil saber quem é quem no bloco

var_global = 10 # Essa variável é global

def multiply(num1, num2):
    var_local = num1 * num2 # Essa variável é local
    print(var_local) # Será impresso na tela o valor da variável var_local
```

```python
# Executando a função multiply

multiply(3,4)
```

```python
# Imprimindo na tela a variável global, que possuí o valor 10

print(var_global)
```

```python
# Imprimindo na tela a variável local, que possuí o valor da multiplicação atribuí a ela
# Notar que irá ocorrer um ero 'name 'var_local' is not defined'
# Esse erro ocorre pois estamos tentando executar uma variável local, que só pode ser utilizada no bloco da função
# Dessa maneira haverá o erro que essa variável não foi declarada

print(var_local)
```

### Criando funções usando outras funções

```python
# Aqui utilizaremos a bilioteca math, que é uma biblioteca de matemática, para diversas realizar diversas operações matemáticas
# Estamos criando uma função chamada numPrimo com paramêtro num
# Utilizamos a função da biblioteca math chamada sqrt, que informa a raíz quadrada de um número
# Assim conseguimos por meio da função abaixo localizar o número primo

import math

def numPrimo(num):
    '''
    Verificando se um número é primo
    '''
    if (num % 2) == 0 and num > 2:
        return print('Este número não é primo')
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if (num % i) == 0:
            return print('Este número não é primo')
    return print('Este número é primo')

# Solicitando a execução da função numPrimo com o paramêtro de um número
numPrimo(540)
```

### Fazendo split dos dados

```python
# Fazendo split dos dados com a função split
# Aqui declaramos um texto na variável chamada texto

texto = 'Esta função será bastante útil para separar grandes volumes de dados'

def split_string(text):
    return text.split(" ")

# Solicitando a execução da função split_string com o paramêtro de um texto, vindo da variável texto declarada acima
split_string(texto)
```

```python
# Podemos atribuir o output(saída) de uma função para uma variável, neste caso, será na variável token

token = split_string(texto)

# Imprimindo o valor da variável token
print(token)
```

```python
# Transformando uma frase em caracteres minusculos com a função lower
# Aqui declaramos um texto na variável chamada caixa_baixa

caixa_baixa = 'Este Texto Deveria Estar Todo Em LowerCase, Ou Seja, Todos Caracateres Em Minisculo'

def lowercase(text):
    return text.lower()

# Solicitando a execução da função lowercase com o paramêtro texto, vindo da variável caixa_baixa
lowercase(caixa_baixa)
```

```python
# Atribuindo o resultado da função lowercase em uma variável
# Imprimindo o resultado da variável

lowercased_string = lowercase(caixa_baixa)

print(lowercased_string)
```

```python
# Funções com número variável de argumentos/paramêtros, ou seja, não será pré-determinado um limite de argumentos/paramêtros
# O comando nessa função chamado vartuple com o (*) antes da variável, quer dizer que a função irá receber diversos argumentos

def printVarInfo(arg1, *vartuple):
    # Imprimindo o valor do primeiro argumentos
    
    print('O parâmetro passado foi: ', arg1)
    
    # Imprimindo o valor do segundo argumento
    for item in vartuple:
        print('O parâmetro passado foi: ', item)
    return;
```

```python
# Fazendo chamada da função para execução somente com um parâmetro

printVarInfo(10)
```

```python
# Fazendo chamada da função para execução com diversos parâmetros

printVarInfo('Chocolate', 'Morango', 'Banana')
```
